# Applying K-means clustering to the processed input data

```python

class K_Means_lib:
  def __init__(self, tweets_lists):
    self.list_of_twitter_tweets = tweets_lists
  
  def k_means(self, k=3, max_iterations=30):

    list_of_centroids = []    
    iter_count = 0
    prev_list_of_centroids = []
    count = 0
    hash_map = dict()
    while count < k:
        random_tweet_idx = rd.randint(0, len(self.list_of_twitter_tweets) - 1)
        if random_tweet_idx not in hash_map:
            count += 1
            hash_map[random_tweet_idx] = True
            list_of_centroids.append(self.list_of_twitter_tweets[random_tweet_idx])

    while (self.has_converged(prev_list_of_centroids, list_of_centroids)) == False and (iter_count < max_iterations):

        #print("running iteration " + str(iter_count))
        cluster_components = self.cluster_assignment(self.list_of_twitter_tweets, list_of_centroids)
        prev_list_of_centroids = list_of_centroids
        list_of_centroids = self.reassign_centroid(cluster_components)
        iter_count = iter_count + 1

    
    sq_sum_err = self.compute_sq_sum_err(cluster_components)

    return cluster_components, sq_sum_err


  def has_converged(self, prev_centroid, new_centroids):

      if len(prev_centroid) != len(new_centroids):
          return False

      for c in range(len(new_centroids)):
          if " ".join(new_centroids[c]) != " ".join(prev_centroid[c]):
              return False

      return True


  def cluster_assignment(self, tweets, centroids):

      clusters = dict()

      for t in range(len(tweets)):
          min_dis = math.inf
          cluster_idx = -1;
          for c in range(len(centroids)):
              dis = self.get_jaccard_dist(centroids[c], tweets[t])
              if centroids[c] == tweets[t]:
                  # print("tweet and centroid are equal with c: " + str(c) + ", t" + str(t))
                  cluster_idx = c
                  min_dis = 0
                  break

              if dis < min_dis:
                  cluster_idx = c
                  min_dis = dis

          if min_dis == 1:
              cluster_idx = rd.randint(0, len(centroids) - 1)

          clusters.setdefault(cluster_idx, []).append([tweets[t]])
          last_tweet_idx = len(clusters.setdefault(cluster_idx, [])) - 1
          clusters.setdefault(cluster_idx, [])[last_tweet_idx].append(min_dis)

      return clusters


  def reassign_centroid(self, cluster_components):

      list_of_centroids = []

      for c in range(len(cluster_components)):
          min_dis_sum = math.inf
          index_of_centroid = -1
          min_dis_dp = []

          for t1 in range(len(cluster_components[c])):
              min_dis_dp.append([])
              dis_sum = 0
              for t2 in range(len(cluster_components[c])):
                  if t1 != t2:
                      if t2 < t1:
                          dis = min_dis_dp[t2][t1]
                      else:
                          dis = self.get_jaccard_dist(cluster_components[c][t1][0], cluster_components[c][t2][0])

                      min_dis_dp[t1].append(dis)
                      dis_sum += dis
                  else:
                      min_dis_dp[t1].append(0)

              if dis_sum < min_dis_sum:
                  min_dis_sum = dis_sum
                  index_of_centroid = t1

          list_of_centroids.append(cluster_components[c][index_of_centroid][0])

      return list_of_centroids


  def get_jaccard_dist(self, tweet_x, tweet_y):

      union = set().union(tweet_x, tweet_y)
      intersect = set(tweet_x).intersection(tweet_y)

      return 1 - (len(intersect) / len(union))


  def compute_sq_sum_err(self, cluster_components):

      sq_sum_err = 0
      for cluster in range(len(cluster_components)):
          for t in range(len(cluster_components[cluster])):
              sq_sum_err = sq_sum_err + (cluster_components[cluster][t][1] * cluster_components[cluster][t][1])

      return sq_sum_err
```

```python
def get_cluster_size(clusters):
  ans_str = ''
  for z in range(0,len(list(clusters.keys()))):
    ans_str += 'Cluster '+str(z)+' : '+str(len(clusters[z]))+' tweets '
  return ans_str
```

```python
exp_k=[]
exp_sq_sum_errors=[]
exp_clusters=[]
for exp in range(1,6):
  clusters_computed, sq_sum_errors = K_Means_lib(df['preprocessed_clean_text'].tolist()).k_means(k=2+exp)
  exp_k.append(2+exp)
  exp_sq_sum_errors.append(sq_sum_errors)
  exp_clusters.append(get_cluster_size(clusters_computed))

results = pd.DataFrame(list(zip(exp_k, exp_sq_sum_errors, exp_clusters)), columns=['K-Values', 'Square Sum Error Values', 'Cluster Sizes'])
```
